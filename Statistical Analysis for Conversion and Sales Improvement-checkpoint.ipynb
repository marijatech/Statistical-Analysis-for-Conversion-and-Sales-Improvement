```python
import numpy as np
import pandas as pd

# Generate data for A/B testing
np.random.seed(42)
visitors_A = 1000
visitors_B = 1200
conversions_A = np.random.binomial(visitors_A, 0.1)  # Assuming 10% conversion rate for A
conversions_B = np.random.binomial(visitors_B, 0.12) # Assuming 12% conversion rate for B

# Generate data for average order values (confidence interval)
avg_order_value = np.random.normal(200, 50, 150)  # Average order value with std dev 50, sample size 150

# Generate data for hypothesis testing (before and after introducing shipping options)
sales_before = np.random.normal(100, 15, 30)  # Average daily sales before
sales_after = np.random.normal(110, 15, 30)  # Average daily sales after

# Display generated data
print(f'Conversions A: {conversions_A}, Conversions B: {conversions_B}')
print(f'Average order values (first 5): {avg_order_value[:5]}')
print(f'Sales before (first 5): {sales_before[:5]}')
print(f'Sales after (first 5): {sales_after[:5]}')
```

```python
from scipy import stats

# A/B testing: Z-test for proportions
conversion_rate_A = conversions_A / visitors_A
conversion_rate_B = conversions_B / visitors_B
pooled_conversion_rate = (conversions_A + conversions_B) / (visitors_A + visitors_B)

# Z-test formula
Z = (conversion_rate_B - conversion_rate_A) / np.sqrt(pooled_conversion_rate * (1 - pooled_conversion_rate) * (1/visitors_A + 1/visitors_B))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

print(f"Z-value: {Z}")
print(f"P-value: {p_value}")
```

```python
import matplotlib.pyplot as plt

# Data for A/B testing
labels = ['Version A', 'Version B']
conversions = [conversions_A, conversions_B]
visitors = [visitors_A, visitors_B]
conversion_rates = [conversions_A / visitors_A * 100, conversions_B / visitors_B * 100]

# Plotting the bar chart
plt.bar(labels, conversion_rates, color=['blue', 'green'])
plt.title('Conversion Rates for A/B Test')
plt.ylabel('Conversion Rate (%)')
plt.show()
```

### Conclusion: A/B Testing

Based on the A/B testing results:

- **Z-value**: 1.92
- **P-value**: 0.055

At a significance level of **\(\alpha = 0.05\)**, the p-value is slightly greater than the threshold, indicating that there is **no statistically significant difference** between the conversion rates of Version A (10%) and Version B (12%).

While Version B showed a higher conversion rate than Version A, this difference is **not statistically significant**. Therefore, there is insufficient evidence to conclude that Version B performs better than Version A in driving conversions. A larger sample size or additional testing might be required for more conclusive results.


```python
#Calculate the mean and standard error of the average order values
mean_order_value = np.mean(avg_order_value)
std_order_value = np.std(avg_order_value, ddof = 1) #Using sample standard deviation
n_orders = len(avg_order_value)

#Z-value for 95% confidence interval
Z_value = stats.norm.ppf(0.975)

#Calculate the margin of error
margin_of_error = Z_value * (std_order_value / np.sqrt(n_orders))

#Confidence interval
CI_lower = mean_order_value - margin_of_error
CI_upper = mean_order_value + margin_of_error

print(f"Mean order value: {mean_order_value}")
print(f"Confidence Interval (95%): ({CI_lower}, {CI_upper})")
```

```python
plt.hist(avg_order_value, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Average Order Value')
plt.xlabel('Order Value ($)')
plt.ylabel('Frequency')
plt.axvline(np.mean(avg_order_value), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.legend()
plt.show()
```

### Conclusion: Confidence Interval for Average Order Value

Based on the data, the mean order value is **195.11** dollars. With a 95% confidence level, the confidence interval for the average order value is between **187.63** and **202.58** dollars.

This means that we are 95% confident that the true average order value lies within this range, which provides insight into customer spending behavior.


```python
# t-test for hypothesis testing
t_stat, p_value = stats.ttest_ind(sales_after, sales_before, equal_var=False)

# Display results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# Combine sales data
sales_data = [sales_before, sales_after]
labels = ['Sales Before', 'Sales After']

# Plotting the boxplot
plt.boxplot(sales_data, tick_labels=labels)  # Use 'tick_labels' instead of 'labels'
plt.title('Sales Before and After New Shipping Options')
plt.ylabel('Sales Amount ($)')
plt.show()

```

### Conclusion: Hypothesis Testing for Sales Before and After

Based on the t-test results:

- **T-statistic**: 2.26
- **P-value**: 0.027

At a significance level of \(\alpha = 0.05\), the p-value is less than the threshold, indicating that we can reject the null hypothesis. Therefore, there is **statistically significant evidence** to suggest that introducing new shipping options has increased the average daily sales.
